KERALA BUS DETAILS

```python
df_0=pd.read_csv("df_route_kerala.csv")
df_0
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException,ElementClickInterceptedException
import time

driver=webdriver.Chrome()

Busname = []
Bustype = []
Departing_time = []
Reaching_time = []
Star_rating = []
Duration = []
Price = []
Seats_available = []
Route_names = []
Route_links = []
State_name=[]

wait = WebDriverWait(driver, 10)

for i, r in df_0.iterrows():
    link = r["Route_link"]
    routes = r["Route_name"]
    states= r["State"] 

    driver.get(link)
    time.sleep(2)  
    
    try:
        elements = driver.find_elements(By.XPATH, f"//a[contains(@href, '{link}')]")
        for element in elements:
            element.click()
            time.sleep(2)  
    except NoSuchElementException:
        pass

    try:
        clicks = driver.find_elements(By.XPATH, "//div[@class='button']")
        for click in clicks:
            # WebDriverWait 
            WebDriverWait(driver, 10).until(EC.element_to_be_clickable(click))           
            ActionChains(driver).move_to_element(click).click().perform()
            time.sleep(2)
    except (NoSuchElementException, ElementClickInterceptedException):
        pass 
    
    scrolling = True
    while scrolling:
        old_page_source = driver.page_source
        ActionChains(driver).send_keys(Keys.END).perform()  
        time.sleep(5)  
        new_page_source = driver.page_source

        if new_page_source == old_page_source:  
            scrolling = False

    busname = driver.find_elements(By.XPATH, "//div[@class='travels lh-24 f-bold d-color']")
    bustype = driver.find_elements(By.XPATH, "//div[@class='bus-type f-12 m-top-16 l-color evBus']")
    departing_time = driver.find_elements(By.XPATH, "//*[@class='dp-time f-19 d-color f-bold']")
    reaching_time = driver.find_elements(By.XPATH, "//*[@class='bp-time f-19 d-color disp-Inline']")
    duration = driver.find_elements(By.XPATH, "//*[@class='dur l-color lh-24']")
    
    try:
        star_rating = driver.find_elements(By.XPATH, "//div[@class='clearfix row-one']/div[@class='column-six p-right-10 w-10 fl']")
    except NoSuchElementException:
        star_rating = [] 
    
    price = driver.find_elements(By.XPATH, '//*[@class="fare d-block"]')
    seats_available = driver.find_elements(By.XPATH, "//div[contains(@class, 'seat-left')]")

    for bus in busname:
        Busname.append(bus.text)
        Route_links.append(link)
        Route_names.append(routes)
        State_name.append(states)
    for bus_type_elem in bustype:
        Bustype.append(bus_type_elem.text)
    for departure in departing_time:
        Departing_time.append(departure.text)
    for reaching in reaching_time:
        Reaching_time.append(reaching.text)
    for duration_elem in duration:
        Duration.append(duration_elem.text)
    for rating_elem in star_rating:
        Star_rating.append(rating_elem.text)
    for price_elem in price:
        Price.append(price_elem.text)
    for seats_elem in seats_available:
        Seats_available.append(seats_elem.text)


print("Successfully Completed")
driver.quit()

data = {
    'Busname': Busname,
    'Bustype': Bustype,
    'Departing_time': Departing_time,
    'Reaching_time': Reaching_time,
    'Duration': Duration,
    'Price': Price,
    'Seats_available': Seats_available,
    'Star_rating': Star_rating,
    'Route_links': Route_links,
    'Route_names': Route_names,
    'State_name': State_name,
}

df_bus_kerala = pd.DataFrame(data)
path = r"C:/Users/DELL/Desktop/RedBus_1/df_bus_kerala.csv"
df_bus_kerala.to_csv(path, index=False)
```
