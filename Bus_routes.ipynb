KERALA ROUTES

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException


driver = webdriver.Chrome()
wait = WebDriverWait(driver, 20)


url_kerala = "https://www.redbus.in/online-booking/ksrtc-kerala/?utm_source=rtchometile"
state_name = "Kerala" 

def KERALA_link_route(state_name, path):
    driver.get(url_kerala)
    time.sleep(2)
    driver.maximize_window()
    
    LINKS_KERALA = []
    ROUTE_KERALA = []
    
    for i in range(1, 3): 
        paths = driver.find_elements(By.XPATH, path)
        
        for link in paths:
            link_url = link.get_attribute("href")
            LINKS_KERALA.append(link_url)
        
        for route in paths:
            ROUTE_KERALA.append(route.text)
        
        try:
            pagination = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]')))         
            next_button = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()="{i + 1}"]')
            
            ActionChains(driver).move_to_element(next_button).perform()
            time.sleep(2)  
            next_button.click()
            time.sleep(3)
        
        except NoSuchElementException:
            print(f"No more pages to paginate at step {i} for {state_name}")
            break
    
    return state_name, LINKS_KERALA, ROUTE_KERALA

state_name, LINKS_KERALA, ROUTE_KERALA = KERALA_link_route(state_name, "//a[@class='route']")
driver.quit()

state_column=[]
for _ in range(len(ROUTE_KERALA)):
    state_column.append(state_name)

df_route_kerala = pd.DataFrame({
    "State": state_column ,
    "Route_name": ROUTE_KERALA,
    "Route_link": LINKS_KERALA
})

path=r"C:/Users/DELL/Desktop/RedBus_1/df_route_kerala.csv"
df_route_kerala.to_csv(path,index=False)

```

```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException


driver = webdriver.Chrome()
wait = WebDriverWait(driver, 20)


url_andhra = "https://www.redbus.in/online-booking/apsrtc/?utm_source=rtchometile"
state_name = "Andhra Pradesh" 

def ANDHRA_PRADESH_link_route(state_name, path):
    driver.get(url_andhra)
    time.sleep(2)
    driver.maximize_window()
    LINKS_ANDHRA_PRADESH=[]
    ROUTE_ANDHRA_PRADESH=[]
     
    for i in range(1,6):
        paths=driver.find_elements(By.XPATH,path)
        
        for links in paths:
            d = links.get_attribute("href")
            LINKS_ANDHRA_PRADESH.append(d)
            
        for route in paths:
            ROUTE_ANDHRA_PRADESH.append(route.text)
        try:
            pagination = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]')))     
            next_button = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()="{i + 1}"]')
            next_button = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()="{i + 1}"]')
            next_button = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()="{i + 1}"]')
            next_button = pagination.find_element(By.XPATH, f'//div[@class="DC_117_pageTabs " and text()="{i + 1}"]')
            
            ActionChains(driver).move_to_element(next_button).perform()
            time.sleep(2)  

            next_button.click()
            time.sleep(3) 

        except NoSuchElementException:
            print(f"No more pages to paginate at step {i} for {state_name}")
            break 
    return state_name,LINKS_ANDHRA_PRADESH,ROUTE_ANDHRA_PRADESH

state_name,LINKS_ANDHRA_PRADESH,ROUTE_ANDHRA_PRADESH=ANDHRA_PRADESH_link_route(state_name,"//a[@class='route']")  
driver.quit()

state_column=[]
for _ in range(len(ROUTE_ANDHRA_PRADESH)):
    state_column.append(state_name)

df_route_andhrapradesh=pd.DataFrame({
    "State": state_column,
    "Route_name":ROUTE_ANDHRA_PRADESH,
    "Route_link":LINKS_ANDHRA_PRADESH})

path=r"C:/Users/DELL/Desktop/REDBUS_1/df_route_andhrapradesh.csv"
df_route_andhrapradesh.to_csv(path,index=False)
```
